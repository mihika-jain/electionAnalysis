Use SMOTE to ass more diabetes patients to the dataset.

```python
from imblearn.over_sampling import SMOTE

X = df_imputed.drop(columns=['diabetes'])
y = df_imputed['diabetes']
# Assume X contains your features and y contains the target variable
smote = SMOTE(sampling_strategy='minority', k_neighbors=3)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Convert back to DataFrame if needed
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled['diabetes'] = y_resampled
```

```python
df_resampled["diabetes"].value_counts()
```

```python
df_resampled = df_resampled.astype(float)
```

Check that features are not correlated and need to be dropped

```python
import pandas as pd
import statsmodels.api as sm

# Prepare your features (X)
X = df_resampled.drop(columns=['diabetes'])

# Add a constant to the model (intercept)
X = sm.add_constant(X)

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns
vif_data['VIF'] = [sm.OLS(X[column], X.loc[:, X.columns != column]).fit().rsquared for column in X.columns]

# Calculate the VIF
vif_data['VIF'] = 1 / (1 - vif_data['VIF'])
```

```python
# Create a matplotlib figure
fig, ax = plt.subplots(figsize=(5, 2))  # Adjust figsize as needed

# Hide the axes
ax.axis('tight')
ax.axis('off')

# Create a table plot of the DataFrame
table = ax.table(cellText=vif_data.iloc[1:].round(2).values, colLabels=vif_data.columns, cellLoc='center', loc='center')

# Bold the first row (column headers)
table.auto_set_font_size(False)
table.set_fontsize(9)

# Set the font weight to 'bold' for the header row
for key, cell in table.get_celld().items():
    if key[0] == 0:  # This is the header row
        cell.set_text_props(fontweight='bold')

# Save the figure as a PNG file
plt.savefig('VIF_image.png', bbox_inches='tight', dpi=300)

# Show the plot in Jupyter (optional)
plt.show()
```

```python
# Identify columns with mean above the threshold
threshold = 10
columns_to_drop = vif_data[vif_data["VIF"] > threshold]
print("Columns to Drop Based on VIF:\n", columns_to_drop)

columns_to_drop = columns_to_drop["Feature"]

# Drop those columns
df_resampled = df_resampled.drop(columns=columns_to_drop.values[1:])
```

See if there is demographic changes between datasets

```python
import pandas as pd
from scipy.stats import chi2_contingency, mannwhitneyu

# Assuming you already have your DataFrames
df1 = diabetes_processed
df2 = diabetes_2017_processed

# Define categorical and continuous demographic variables
categorical_vars = ['race', 'region', 'sex', 'marital_status']
continuous_vars = ['age','height','weight', 'bmi']

# Chi-squared tests for categorical variables
for var in categorical_vars:
    contingency_table = pd.crosstab(df1[var], df2[var])
    chi2, p, _, _ = chi2_contingency(contingency_table)
    print(f'Chi-squared test for {var}: p-value = {p}')

# Mann-Whitney U test for continuous variables
for c in continuous_vars:
    c1 = df1[c].dropna()  # Drop NaN values if any
    c2 = df2[c].dropna()
    
    # Perform the Mann-Whitney U test
    u_stat, p_c = mannwhitneyu(c1, c2)
    print(f'{c} comparison p-value = {p_c}')
```
