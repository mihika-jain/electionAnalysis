```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Example DataFrame with multiple categories and missing values

# Step 1: Impute missing values with the most frequent category (mode)
cats_simpleImpute = ['region', 'marital_status']
simple_imputer = SimpleImputer(strategy='most_frequent')
df_imputed = diabetes_processed
df_imputed[cats_simpleImpute] = simple_imputer.fit_transform(diabetes_processed[cats_simpleImpute])

# Reload the module to reflect any updates
importlib.reload(prepare_data)

df_imputed = prepare_data.categories_oneHot(df_imputed, ["region"], True, 
{1: 'Northwest', 2: 'Midwest', 3: 'South', 4: 'West'})

df_imputed = prepare_data.categories_oneHot(df_imputed, ["marital_status"], True,
    { 0: 'Never_married', 1: 'Married', 2: 'Married', 3: 'Married', 4: 'Widdowed', 5: 'Divorced', 6: 'Separated', 7: 'Never_married', 8: 'Living with partner'})

```

Impute using KNN Imputer

```python
from sklearn.impute import KNNImputer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

# Initialize the KNNImputer with a specified number of neighbors (e.g., 5)
KNN_imputer = KNNImputer(n_neighbors=5)

# Fit the imputer to the data and transform it
df_imputed = pd.DataFrame(KNN_imputer.fit_transform(df_imputed), columns=df_imputed.columns)

# Print the imputed dataset
print(df_imputed)
```

```python
#round binary columns
for c in binary_cols:
    df_imputed[c] = df_imputed[c].round(0)
```

```python
#checking round opperation worked
for c in df_imputed.columns:
    print(df_imputed[c].value_counts())
    
```

```python
df_imputed["diabetes"].value_counts()
```

```python
from imblearn.over_sampling import SMOTE

X = df_imputed.drop(columns=['diabetes'])
y = df_imputed['diabetes']
# Assume X contains your features and y contains the target variable
smote = SMOTE(sampling_strategy=0.5, k_neighbors=3)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Convert back to DataFrame if needed
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled['diabetes'] = y_resampled
```

```python
df_resampled.describe()
```

```python
df_resampled["diabetes"].value_counts()
```

```python
import pandas as pd
import statsmodels.api as sm

# Prepare your features (X)
# Assuming 'diabetes' is your target variable and all other columns are features
X = df_resampled.drop(columns=['diabetes'])

# Add a constant to the model (intercept)
X = sm.add_constant(X)

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns
vif_data['VIF'] = [sm.OLS(X[column], X.loc[:, X.columns != column]).fit().rsquared for column in X.columns]

# Calculate the VIF
vif_data['VIF'] = 1 / (1 - vif_data['VIF'])

# Display the VIF results
print(vif_data)
```

```python
# Identify columns with mean above the threshold
threshold = 10
columns_to_drop = vif_data[vif_data["VIF"] > threshold]
print(columns_to_drop)
#print(vif_data[vif_data["VIF"] > threshold])
columns_to_drop = columns_to_drop["Feature"]
print(columns_to_drop.values)

# Drop those columns
df_resampled = df_resampled.drop(columns=columns_to_drop.values[1:])
```

```python
from functions import prepare_data

# Reload the module to reflect any updates
importlib.reload(prepare_data)

train_data, val_data, test_data = prepare_data.split_data(df_resampled, 0.8, 0.1, 0.1)
```

```python
# Save the datasets to CSV files
train_data.to_csv('train_data.csv', index=False)
val_data.to_csv('val_data.csv', index=False)
test_data.to_csv('test_data.csv', index=False)
```
